```python
import pandas as pd

# Step 1: Load the chlorophyll time series CSV file
file_path = r"C:\Users\grace.UOPNET\Desktop\chlor_a.csv"
df = pd.read_csv(file_path)

# Preview the first few rows
print(df.head())
```

```python
# Step 2: Convert 'date' column to datetime and set as index
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Confirm it worked
print(df.index)
print(df[['mean']].head())
```

```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt
plt.style.use('grayscale')  # Clean style with grid and contrast


# Step 4: Plot main line
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['mean'], color='royalblue', linewidth=2, label='Monthly Mean')

# Highlight Mays
may_data = df[df.index.month == 5]
plt.scatter(may_data.index, may_data['mean'], color='darkorange', s=60, label='May Highlight')

# Add red star for May 2020
may2020 = df.loc['2020-05-01']
plt.plot(may2020.name, may2020['mean'], marker='*', color='red', markersize=15, label='May 2020 (Amphan)')


# Labels and formatting
plt.title("Monthly Chlorophyll-a Concentration (2014–2024)", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Mean Chlorophyll-a (mg/m³)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('chlorophyll_timeseries.png', dpi=300)

plt.show()

```
